```python
from econml.dml import CausalForestDML
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split



X_train, X_test, T_train, T_test, Y_train, Y_test = train_test_split(X, T, Y, test_size=0.2, random_state=42)



```

```python
print(X_train.shape)
print(Y_train.shape)
print(T_train.shape)
```

```python
print(set(T_train))  # Check if T_train has discrete values
```

```python
T_train = T_train.map({0: 0, 1: 1, 4: 1, 5: 1})
```

```python
from econml.dml import LinearDML

causal_model = LinearDML(discrete_treatment=True)
causal_model.fit(Y_train, T_train, X=X_train)
```

The above is a simple OLS model to check if the data is good enough to fit the necessities as in the Causal Forest due to the computational power required my colab was constantly crashing


```python
causal_forest = CausalForestDML(model_t=RandomForestRegressor(),
                                model_y=RandomForestRegressor(),
                                discrete_treatment=True)

causal_forest.fit(Y=Y_train, T=T_train, X=X_train)
```

Minimised the records that will be using to test how the model performs and tried to replicate the results by sampling of the overall population for now only took the first 1000 as stratified sampling was also very computationally extensive

```python
treatment_effects = causal_forest.effect(X_test)
print(f"Average Treatment Effect (ATE): {treatment_effects.mean()}")
```

```python
treatment_effects = causal_forest.effect(X_test)
print(f"Average Treatment Effect (ATE): {treatment_effects.mean()}")
```

The Average Treatment Effect (ATE) = 0.047 (4.7%) suggests that on average, users exposed to different banner_pos treatments experience an 4.7% increase in click rate compared to the baseline, indicating positive causal affect.

This means that changing the app (banner_pos) has a causal impact on click rate and engagement.

Heterogeneous Treatment Effects (HTE) being uniform indicates that this 4.7% increase is consistent across different user subgroups, meaning all users respond similarly to banner position changes.

After checking the Heterogenous Treatment Affect and visualising it, it is evident that the histogram is uniform and is not heavily skewed indicating that all the covariates are getting similar treatment affect. This can be an indication of homogenous data.


Based on all the ATE that we have observed we got the maximum positive causal affect rate for different banner_pos that is 4.7 percent, which is something interesting as we have worked on the anomilies of the banner_position frequencies and the click rates they were receiving. This is then followed by the site_domain and the app_id ATE so a good business decision to get the maximum engagement would be to find the best banner_pos firstly based on the site_domain as we saw during the EDA there were banner_pos which had been used less but got more number of clicks might be due to the site_domain.

A/B dynamic testing can be done for different apps based on the app_id that have maximum frequencies with different banner_pos and site_domains to get maximum engagement. Additionally, the interaction effect of banner position and app (0.8%) shows that while the combined impact is smaller, certain apps may still benefit from specific ad placements.
